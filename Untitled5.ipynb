```python
class Bank_v1():
    Bname='SBI'
    NOB=123
    ROI=0.07
    
    def __init__(self,Name,MNo,AccNo,Bal,Pin):
        self.Name=Name
        self.MNo=MNo
        self.AccNo=AccNo
        self.Bal=Bal
        self.Pin=Pin
        
    def Details(self):
        print(f'NAME       : {self.Name}')
        print(f'MOBILE NO  :{self.MNo}')
        print(f'ACCOUNT NO :{self.AccNo}')
        print(f'BALANCE    :{self.Bal}')
        print(f'PIN        :{self.Pin}')
        
    def checkBal(self):
        if self.Pin==self.Getpin():
            print(f'Available Balance is {self.Bal}')
        else:
            print('incorrect pin')
            
    def withDraw(self):
        if self.Pin==self.Getpin():
            amt=self.Getamt()
            if self.Bal>=amt:
                self.Bal-=amt
                print(f'{amt} credited successfully')
                print(f'Available Balance is {self.Bal}')
            else:
                print('insufficient Balance')
        else:
            print('incorrect pin')
            
            
    def Deposit(self):
        amt=self.Getamt()
        self.Bal+=amt
        print(f'{amt} depositted successfully')
        print(f'Available Balance is {self.Bal}')
        
    @staticmethod
    def Getpin():
        return int(input('Enter 4 dig pin : '))
    
    @staticmethod
    def Getamt():
        return int(input('Enter Amount : '))
    

class Bank_v2(Bank_v1):
    def __init__(self,Name,MNo,AccNo,Bal,Pin,Mail,Pan):
        super().__init__(Name,MNo,AccNo,Bal,Pin)
        self.Mail=Mail
        self.Pan=Pan
        
    def Details(self):
        Bank_v1.Details(self)
        print(f'MAIL :{self.Mail}')
        print(f'PAN  :{self.Pan}')
        
class Bank_v3(Bank_v2):
    def __init__(self,Name,MNo,AccNo,Bal,Pin,Mail,Pan,Aadhar,DOB):
        Bank_v2.__init__(self,Name,MNo,AccNo,Bal,Pin,Mail,Pan)
        self.Aadhar=Aadhar
        self.DOB=DOB
            
    def Details(self):
        Bank_v2.Details(self)
        print(f'Aadhar No:{self.Aadhar}')
        print(f'DOB      :{self.DOB}')
        
        
ob1=Bank_v3('ram',1234567,'ab124j42vd',200000,2222,'ram@gmail.com',12435564,32326898,'11-9-2000')
ob2=Bank_v3('sam',1234769,'ab124j4acfr',300000,1111,'sam@gmail.com',12435432,12123456,'12-3-2001')
ob1.checkBal()
ob2.withDraw()
ob1.Deposit()
                
        
                
```
